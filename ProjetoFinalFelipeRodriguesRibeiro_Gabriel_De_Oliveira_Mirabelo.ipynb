Funções Compostas e Operações com funções

Nosso trabalho tem como base criar um padrão para que sejam feitas funções compostas e operações com as mesmas.


```python
def função(x): #Calcula as compostas das funções abaixo a partir de qualquer valor dado ao domínio x.
 
 f = 2*x - 4  # função f(x).

 g = 4*x + 4 # função g(x).

 h = 10*x # função h(x).

 fof = 2*f - 4 #Dada uma função f (f: x → f(x) ), e a repetição da mesma, se tem a composta de f com f, que é representada por fof.
 print(f'fof = {fof}')

 fog = 2*g - 4 #Dada uma função f (f: x → f(x)) e uma função g (g: x → G(x)),a função composta de f com g é representada por fog.
 print(f'fog={fog}')

 foh = 2*h - 4 #Dada uma função f (f: x → f(x)) e uma função h (h: x → h(x)),a função composta de f com h é representada por foh.

 gog = 4*g + 4 #Dada uma função g (g: x → g(x)), e a repetição da mesma, se tem a composta de g com g, que é representada por gog.
 print(f'gog = {gog}')

 gof = 4*f + 4 #Dada uma função g (g: x → g(x)) e uma função f (f: x → f(x)),a função composta de g com f é representada por gof.
 print(f'gof= {gof}')

 goh = 4*h +4  #Dada uma função g (g: x → g(x)) e uma função h (h: x → h(x)),a função composta de g com  h representada por goh.
 print(f'goh= {goh}')
 
 hoh = 10*h #Dada uma função h (f: x → h(x) ), e a repetição da mesma, se tem a composta de h com h, que é representada por hoh.
 print(f'hoh= {hoh}')

 hof = 10*f #Dada uma função h (h: x → h(x)), e uma função f (f: x → f(x)),a função composta de h com f é representada por hof.
 print(f'hof= {hof}')

 hog = 10*g #Dada uma função h (h: x → h(x)), e uma função g (g: x → G(x)),a função composta de h com g é representada por hog.
 print(f'hog= {hog}')

 fogoh = 4*goh - 4 #Dada uma função f (f: x → f(x)), uma função g(g: x → g(x)) e uma função h (h: x → h(x)),se tem f(f: x → g(h(x))), a função composta de f com g(h(x)) é representada por fogoh.
 print(f'fogoh= {fogoh}')
 
 plt.figure()
 plt.plot(x,f, marker = '.', color = 'Green', linestyle = '-')
 plt.title('Gráfico da composta de f(x)')
 plt.xlabel('x')
 plt.ylabel('F(x)')
 plt.show()
#Gráfico da Função F(x)

 plt.figure()
 plt.plot(x,g, marker = '.', color = 'Red', linestyle = '-')
 plt.title('Gráfico da g(x)')
 plt.xlabel('x')
 plt.ylabel('g(x)')
 plt.show()
#Gráfico da Função G(x)

 plt.figure()
 plt.plot(x,h, marker = '.', color = 'Grey', linestyle = '-')
 plt.title('Gráfico da h(x)')
 plt.xlabel('x')
 plt.ylabel('h(x)')
 plt.show()
#Gráfico da Função H(x) 

 plt.figure()
 plt.plot(x,fof, marker = '.', color = 'Blue', linestyle = '-')
 plt.title('Gráfico da composta fof(x)')
 plt.xlabel('x')
 plt.ylabel('fof(x)')
 plt.show()
#Gráfico da Função f(f(x))

 plt.figure()
 plt.plot(x,fog, marker = '.', color = 'Yellow', linestyle = '-')
 plt.title('Gráfico da composta fog(x)')
 plt.xlabel('x')
 plt.ylabel('fog(x)')
 plt.show()
#Gráfico da Função f(g(x))

 plt.figure()
 plt.plot(x,foh, marker = '.', color = 'Pink', linestyle = '-')
 plt.title('Gráfico da composta foh(x)')
 plt.xlabel('x')
 plt.ylabel('foh(x)')
 plt.show()
#Gráfico da Função f(h(x))

 plt.figure()
 plt.plot(x,gof, marker = '.', color = 'Black', linestyle = '-')
 plt.title('Gráfico da composta gof(x)')
 plt.xlabel('x')
 plt.ylabel('gof(x)')
 plt.show()
#Gráfico da Função g(f(x))

 plt.figure()
 plt.plot(x,gog, marker = '.', color = 'Green', linestyle = '-')
 plt.title('Gráfico da composta de gog(x)')
 plt.xlabel('x')
 plt.ylabel('gog(x)')
 plt.show()
#Gráfico da Função g(g(x))

 plt.figure()
 plt.plot(x,gof, marker = '.', color = 'Brown', linestyle = '-')
 plt.title('Gráfico da composta de gof(x)')
 plt.xlabel('x')
 plt.ylabel('gof(x)')
 plt.show()
#Gráfico da Função g(f(x))
 
 plt.figure()
 plt.plot(x,goh, marker = '.', color = 'Green', linestyle = '-')
 plt.title('Gráfico da composta de goh(x)')
 plt.xlabel('x')
 plt.ylabel('goh(x)')
 plt.show()
#Gráfico da Função g(h(x))

 plt.figure()
 plt.plot(x,hoh, marker = '.', color = 'Red', linestyle = '-')
 plt.title('Gráfico da composta de hoh(x)')
 plt.xlabel('x')
 plt.ylabel('hoh(x)')
 plt.show()
#Gráfico da Função h(h(x))

 plt.figure()
 plt.plot(x,hof, marker = '.', color = 'Black', linestyle = '-')
 plt.title('Gráfico da composta de hof(x)')
 plt.xlabel('x')
 plt.ylabel('hof(x)')
 plt.show()
 #Gráfico da Função h(f(x))

 plt.figure()
 plt.plot(x,hog, marker = '.', color = 'Yellow', linestyle = '-')
 plt.title('Gráfico da composta de hog(x)')
 plt.xlabel('x')
 plt.ylabel('hog(x)')
 plt.show()
 #Gráfico da Função h(g(x))

 plt.figure()
 plt.plot(x,fogoh, marker = '.', color = 'Blue', linestyle = '-')
 plt.title('Gráfico da composta de fogoh(x)')
 plt.xlabel('x')
 plt.ylabel('fogoh(x)')
 plt.show()
 #Gráfico da Função f(g(h(x))

 return x
#obs: esse é um molde para se achar compostas e suas  variações (Fog/Foh/gof/goh/hof/hog/hoh/gog/fof/fogoh)
```

```python
x = np.arange(-20,20)
def Composta(g): #dado uma função f, a composta fog pode ser definida para qualquer g no domínio definido.
 y = g
 #f = 3x**3 + 12x**2 + 2x
 fog = 3*(y**3)+12*(y**2)+ 2*y
 print(f'fog={fog}')

 plt.figure()
 plt.plot(x,fog, marker = 'd',color = 'Blue')
 plt.xlabel('x')
 plt.ylabel('FoG(x)')
 plt.show()
```

Operações com funções

```python
x = np.arange(0,10) #domínio.
def operações(a,b): #calcula as operações para qualquer função f e g, que são definidas a partir de a e b.

 f = a #contradominio.
 g = b #contradominio.

 Soma = f + g #soma dos conjuntos a & b.
 print(f'f+g = {Soma}')

 plt.figure()
 plt.plot(x,Soma, marker = '.', color = 'Green', linestyle = '-')
 plt.title('Gráfico de f + g')
 plt.xlabel('x')
 plt.ylabel('f(x) + g(x)')
 plt.show()
#Gráfico da soma de f(x)+g(x)

 Subtração = f - g #subtração dos conjuntos a & b.
 print(f'f-g= {Subtração}')

 plt.figure()
 plt.plot(x,Subtração, marker = '.', color = 'Blue', linestyle = '-')
 plt.title('Gráfico de f - g')
 plt.xlabel('x')
 plt.ylabel('f(x) - g(x)')
 plt.show()
 #Gráfico da subtração de f(x) - g(x)

 Produto = f*g #produto do conjunto a em relação com b.
 print(f'f*g = {Produto}')

 plt.figure()
 plt.plot(x,Produto, marker = '.', color = 'Red', linestyle = '-')
 plt.title('Gráfico de f*g')
 plt.xlabel('x')
 plt.ylabel('f(x) * g(x)')
 plt.show()
#Gráfico do produto de f(x) * g(x)

 Divisão= f/g #divisão do conjunto a em relação com b.
 print(f'f/g = {Divisão}')

 plt.figure()
 plt.plot(x,Divisão, marker = '.', color = 'Black', linestyle = '-')
 plt.title('Gráfico de f/g')
 plt.xlabel('x')
 plt.ylabel('f(x) / g(x)')
 plt.show()
#Gráfico da divisão de f(x) / g(x)

 Divisão2= g/f ##produto do conjunto b em relação com a.
 print(f'g/f = {Divisão2}')

 plt.figure()
 plt.plot(x,Divisão2, marker = '.', color = 'Yellow', linestyle = '-')
 plt.title('Gráfico de g/f')
 plt.xlabel('x')
 plt.ylabel('g(x) / f(x)')
 plt.show()
##Gráfico da divisão de g(x) / f(x)

#obs: foi criado um molde para calcular operações com funções.
```
