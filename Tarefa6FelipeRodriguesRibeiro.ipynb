```python
import matplotlib.pyplot as plt
```

Questão 01

```python
t = np.arange(0, 10, 0.05) #tempo de 0s a 10s com incremento de 0.05
print(t)
p = np.pi
print(p)
```

```python
v = np.exp(-0.5*t)*(np.cos((2*p)*(0.8*t))) #fórmula da tensão
plt.figure(figsize=(20,5))
plt.plot(t,v, color= 'green', marker = 'o', linestyle= 'dotted')
plt.title('Gráfico da tensão em função do tempo')
plt.xlabel('tempo (s)')
plt.ylabel('tensão (V)')
plt.show()
```

Questão 02

```python
c = np.linspace(-20,100,100)
print(c)
f = (9/5*c) + 32
plt.figure(figsize=(12,6))
plt.title('Gráfico da equivalência da temperatura em Fahrenheits em função da temperatura em Celsius')
plt.plot(c,f, color= 'blue', marker = 'o' )
plt.xlabel('graus celsius(°c)')
plt.ylabel('graus fahrenheit(°F)')
plt.show()
```
