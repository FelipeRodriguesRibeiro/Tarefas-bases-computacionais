```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Questão 01

```python
x = pd.read_csv('https://raw.githubusercontent.com/rnwatanabe/BasesComputacionais/0cb190ba78cfac056fc492790c1043de4ef4e9c2/aula9/dados/BDSinfo.csv', sep = '\t')
x
```

```python
plt.figure()
plt.plot(x['Height'],x['FootLen'],marker='o', color='red', linestyle='')
plt.xlabel('Altura')
plt.ylabel('Tamanho do pé')
plt.title('Gráfico de dispersão')
plt.grid()
plt.show()
```

```python
corr_tamanhodopé_altura = x['FootLen'].corr(x['Height'])
print(corr_tamanhodopé_altura)
```

```python
l = x['Height']
n = x['FootLen']
mediaX = np.mean(l)
mediaY = np.mean(n)
desvioX = l - mediaX
desvioY = n - mediaY
m = np.sum(desvioX*desvioY)/np.sum(desvioX**2)
b = mediaY - m*mediaX
print(m)
print(b)
```

```python
plt.figure()
plt.plot(l , n, marker='o', color='red', linestyle='')
plt.plot(l, m*l+b,linestyle = '-', color = 'blue')
plt.xlabel('Altura')
plt.ylabel('Tamanho do pé')
plt.title('reta de regressão')    
plt.grid()
plt.show()
```

```python
p = np.polyfit(x['Height'], x['FootLen'],deg=1)
plt.figure()
plt.plot(x['Height'], x['FootLen'], marker='o', color='red', linestyle='')
plt.plot(x['Height'], np.polyval(p, x['Height']),marker='', color='blue', linestyle='-')
plt.xlabel('altura')
plt.ylabel('tamanho do pé')
plt.show()
```

Questão 03

```python
dados = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/f2740361de0c972edfa8f878935d282025732213/dados/BDSinfo.csv',sep = '\t')
dados
```

```python
plt.figure()
plt.plot(dados['Age'],dados['Nmedication'], marker='o', color='red', linestyle='')
plt.xlabel('Idade')
plt.ylabel('número de medicação')
plt.title('Gráfico de dispersão')
plt.grid()
plt.show()
```

```python
corr_númerodemedicamentos_idade = dados['Nmedication'].corr(dados['Age'])
print(corr_númerodemedicamentos_idade)
```

```python
p = np.polyfit(dados['Age'], dados['Nmedication'],deg=2)
plt.figure()
plt.plot(dados['Age'],dados['Nmedication'],marker='o', color='red', linestyle='')
plt.plot(dados['Age'].sort_values(), np.polyval(p, dados['Age'].sort_values()), marker='', color='blue', linestyle='-')
plt.xlabel('Idade (anos)')
plt.ylabel('número de medicamentos')
plt.title('Reta de regressão')
plt.show()
```
