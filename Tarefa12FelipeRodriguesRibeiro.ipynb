```python
v = np.zeros(251)
v[0] = 10
for i in range(0,251,1):
 v[i] = v[i-1]+0.01*(-9.81)
 print(f'i={i},v={v[i]}')
```

```python
y = np.zeros(251)
y[0] = 0
for i in range(0,251,1):
 y[i]= y[i-1]+0.01*v[i-1]
 print(f'i={i},y={y[i]}')
```

```python
t = np.zeros(251)
t[0] = 0
for i in range(1,251):
  t[i] = t[i-1] + 0.01
  print(f'i={i},t[i]={t[i]}')
```

```python
plt.figure()
plt.plot(y,t)
plt.xlabel('y')
plt.ylabel('t')
plt.show()
```

```python
N = 10
y = 0
t = np.arange(0,10,0.000001)
for i in range(1,N+1,1):
  y = y + (np.sin(2*i-1)*t)/2*i-1
x = 4/np.pi * y
print(f't={t},x={x}')
```

```python
plt.figure()
plt.plot(x,t,marker = 'o')
plt.xlabel('x')
plt.ylabel('t')
plt.show()
```
