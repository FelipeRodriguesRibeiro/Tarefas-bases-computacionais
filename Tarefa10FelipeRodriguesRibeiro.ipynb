Questão 02

```python
taxa = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/inflacaoMensal.csv')
taxa
```

```python
taxa['Mês/Ano'] = taxa['Mês']*1/12/taxa['Ano']
taxa['Mês/Ano']
```

```python
plt.figure()
plt.plot(taxa['Inflação'],taxa['Mês'], marker = 'o')
plt.xlabel('taxa de Inflação')
plt.ylabel('meses')
plt.show()
```

```python
taxa.groupby('Ano')['Inflação'].max()
```

```python
taxa.groupby('Mês')['Inflação'].max()
```

Questão 03

```python
Netflix = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/netflix.csv')
Netflix
Netflix['Categoria'] = 'Filme'  
Netflix['Categoria'][Netflix['Title'].str.contains(": Temporada|: Stranger|: Parte")] = 'Série'  
Netflix['Programa'] = Netflix['Title']  
Netflix[['Programa','Episódio']] = Netflix[Netflix['Categoria']=='Série']['Title'].str.split(pat = ': Temporada|: Stranger Things|: Parte', expand = True, n = 1)   
Netflix.loc[Netflix['Categoria']=='Filme', 'Programa'] = Netflix.loc[Netflix['Categoria']=='Filme', 'Title']  
Netflix = Netflix.drop(columns = 'Title')
Netflix
```

```python
mostview= Netflix.sort_values(['Programa'],ascending=False)
mostview
```

```python
mostview.head(10)
```

```python
Netflix.groupby('Date')['Programa'].count()
```
